```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y(x):
    return np.sin(x / 5) * np.log10(6 * x) + 3

def cos(x):
    return np.cos(x)
```

```python
xs = np.arange(1, 16, dtype=float)
print("xs = ", xs)
```

```python
ys = []
for x in xs:
    ys.append(cos(x))
print(ys)
```

```python
plt.plot(xs, ys, label="y(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Аппроксимация по 2 точкам

```python
A1 = [[1, xs[0]],
      [1, xs[-1]]]
b1 = [ys[0], ys[-1]]

ws1 = np.linalg.solve(A1, b1)
print(ws1)
```

```python
ys_sol1 = []
for x in xs:
    ys_sol1.append(ws1[0] + ws1[1] * x)
print(ys_sol1)
```

```python
plt.plot(xs, ys, label="y(x)")
plt.plot(xs, ys_sol1, label="sol1")
plt.legend()
plt.show()
```

## Аппроксимация по 3 точкам

```python
A2 = [[1.0, xs[0], xs[0] ** 2],
      [1.0, xs[len(xs) // 2], xs[len(xs) // 2] ** 2],
      [1.0, xs[-1], xs[-1] ** 2]]
b2 = [ys[0], ys[len(ys) // 2], ys[-1]]

ws2 = np.linalg.solve(A2, b2)
print(ws2)
```

```python
ys_sol2 = []
for x in xs:
    ys_sol2.append(ws2[0] + ws2[1] * x + ws2[2] * x ** 2)
print(ys_sol2)
```

```python
plt.plot(xs, ys, label="y(x)")
plt.plot(xs, ys_sol2, label="sol2")
plt.legend()
plt.show()
```

## Аппроксимация по 4 точкам

```python
A3 = [[1.0, xs[0], xs[0] ** 2, xs[0] ** 3],
      [1.0, xs[4], xs[4] ** 2, xs[4] ** 3],
      [1.0, xs[10], xs[10] ** 2, xs[10] ** 3],
      [1.0, xs[-1], xs[-1] ** 2, xs[-1] ** 3]]
b3 = [ys[0], ys[4], ys[10], ys[-1]]

ws3 = np.linalg.solve(A3, b3)
print(ws3)
```

```python
ys_sol3 = []
for x in xs:
    ys_sol3.append(ws3[0] + ws3[1] * x + ws3[2] * x ** 2 + ws3[3] * x ** 3)
print(ys_sol3)
```

```python
plt.plot(xs, ys, label="y(x)")
plt.plot(xs, ys_sol3, label="sol3")
plt.legend()
plt.show()
```
